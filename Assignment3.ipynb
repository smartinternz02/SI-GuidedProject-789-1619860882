```python
import seaborn as sns 
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x="petal_length",y="petal_width",data=iris)
```

```python
sns.lineplot(x="petal_length",y="petal_width",data=iris)
```

```python
iris.plot(kind='scatter',x="petal_length",y='petal_width')
```

```python
plt.scatter(iris["petal_length"],iris['petal_width'])
```

```python
sns.barplot(x="petal_length",y="petal_width",data=iris)
```

```python
sns.boxplot(x="petal_length",y="petal_width",data=iris)
```

```python
sns.catplot(x="petal_length",y="petal_width",data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot=True)
```
